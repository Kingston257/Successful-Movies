```python
-- Top Rated Movies
SELECT * FROM "Movies.csv" 
ORDER BY "Vote Average" DESC 
LIMIT 5;
```

```python
-- Least Rated Movies
SELECT * FROM "Movies.csv" 
ORDER BY "Vote Average" 
LIMIT 5;
```

```python
-- Highest Grossing Movies
SELECT * FROM "Movies.csv" 
ORDER BY "Revenue" DESC 
LIMIT 5;
```

```python
-- Movies that lost money
SELECT * FROM "Movies.csv"
WHERE "Revenue" < "Budget";
```

```python
-- ^^ sorted by Ratings
SELECT * FROM "Movies.csv"
WHERE "Revenue" < "Budget"
ORDER BY "Vote Average" DESC
```
